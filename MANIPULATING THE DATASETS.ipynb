# APPLYING THE FUNCTIONS()

```python
# doubliing the records in the temp_max column using lamda function
f=lambda x: x*2
df["temp_max"]=df["temp_max"].apply(f)
df
```

# SORTING

```python
# sorting values in the precipitation column in ascending order
df.sort_values(by="precipitation")
```

```python
# sorting values in the precipitation column in descending order
df.sort_values(by="precipitation",ascending=False)
df
```

```python
# filtering values greater than 50 in the temp_max column
df["temp_max"]>50
```

```python
# filtering values greater than 50 in the temp_max column
filter1=df["temp_max"]>50
# Applying the filtered data into the dataframe
filter_new=df[filter1]
filter_new
```

```python
# filtering values greater than 50 in the temp_max column
filter1=(df["temp_max"]>50) & (df["precipitation"]<1)
# Applying the filtered data into the dataframe
filter_new=df[filter1]
filter_new
```
